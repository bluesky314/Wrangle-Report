```python
# Combine dog states columns into one column
columnsmelt = ['doggo', 'floofer', 'pupper', 'puppo']

columns=[]
for e in fin_df.columns.tolist():
    if e not in columnsmelt:
        columns.append(e)
        
# Melt the the columns into values
fin_df = pd.melt(fin_df, id_vars = columns, value_vars = columnsmelt, 
                         var_name = 'stages', value_name = 'dog_stage')
```

```python
#Convert float variables (in_reply_to_status_id, in_reply_to_user_id, 
#retweeted_status_id, retweeted_status_user_id) to object.
df_mas.in_reply_to_status_id_x = df_mas.in_reply_to_status_id_x.astype(str)
df_mas.in_reply_to_user_id_x = df_mas.in_reply_to_user_id_x.astype(str)
df_mas.retweeted_status_id = df_mas.retweeted_status_id.astype(str)
df_mas.retweeted_status_user_id = df_mas.retweeted_status_user_id.astype(str)
```

I want to convert these to the correct types so that if we perform a .describe then we don't want additional columns that cloud our main data

```python
#convert timestamp to date time format
df_mas.timestamp = pd.to_datetime(df_mas.timestamp)
df_mas.retweeted_status_timestamp = pd.to_datetime(df_mas.retweeted_status_timestamp)
```

Probably one of the most useful metric is date-time and so we need this to see trends over time in the correct format

```python
#Change missing values in 'name' from 'None' to NaN 
df_mas['name'] = df_mas['name'].replace('None', np.NaN)
```

```python
#new column ratings numeric
df_mas['rating_numerator'] = df_mas['rating_numerator'].astype(float)
df_mas['rating_denominator'] = df_mas['rating_denominator'].astype(float)
```

```python
df_mas['rating_numeric']=df_mas.rating_numerator/df_mas.rating_denominator
```

This new column has the ratings in numeric format to perform calculations on

```python
#new column ratings string
df_mas['rating_numerator'] = df_mas['rating_numerator'].astype(str)
df_mas['rating_denominator'] = df_mas['rating_denominator'].astype(str)
```

```python
df_mas['rating']=df_mas.rating_numerator +'/' +df_mas.rating_denominator
```

This new column has the ratings in string format to preserve the visual aesthetic on the twitter page

```python
#Make p1,p2,p3 all caps
df_mas['p1'] = map(lambda x: x.upper(), df_mas['p1'])
df_mas['p2'] = map(lambda x: x.upper(), df_mas['p2'])
df_mas['p3'] = map(lambda x: x.upper(), df_mas['p3'])
```
